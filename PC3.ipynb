```python
import numpy as np
import random as rd
import matplotlib.pyplot as plt
```

```python
nc, nf = 100,100
M = np.full((nc,nf,3), [255,255,255])

i = rd.randint(0,nf-1)
j = rd.randint(0,nc-1)

rr = rd.randint(0,255)
rg = rd.randint(0,255)
rb = rd.randint(0,255)
color = [rr,rg,rb]
M[i,j] = color

######################################

#FILA 
for y in range(0,j+1,1):
   M[i,y] = color

for y in range(j,nc,1):
   M[i,y] = color

#COLUMNA
for x in range(0,i,1):
   M[x,j] = color

for x in range(i,nf,1):
   M[x,j] = color


#DIAGONAL PRINCIPAL
for x,y in zip(range(i,-1,-1),range(j,-1,-1)):
   M[x,y] = color

for x,y in zip(range(i,nf,1),range(j,nc,1)):
   M[x,y] = color

#DIAGONAL SEGUNDARIA
for x,y in zip(range(i,nf,1),range(j,-1,-1)):
   M[x,y] = color
   
for x,y in zip(range(i,-1,-1),range(j,nc,1)):
   M[x,y] = color


plt.imshow(M)
plt.title("GRÁFICA DE COLUMNA, FILA Y DIAGONALES")
plt.show()
print(f"PUNTO ALEATORIO GENERADO EN ({i}, {j})")

###################################################

M = np.full((nc,nf,3), [255,255,255])
#RAMA 1-2
for x,y in zip(range(i,-1,-1),range(j,-1,-1)):
   M[x,y] = color
for x,y in zip(range(i,-1,-1),range(j,nc,1)):
   M[x,y] = color


plt.imshow(M)
plt.title("GRÁFICA DE RAMA 1-2")
plt.show()
print(f"PUNTO ALEATORIO GENERADO EN ({i}, {j})")


###################################################

M = np.full((nc,nf,3), [255,255,255])

#RAMA 2-3
for x,y in zip(range(i,nf,1),range(j,-1,-1)):
   M[x,y] = color
for x,y in zip(range(i,-1,-1),range(j,-1,-1)):
   M[x,y] = color

plt.imshow(M)
plt.title("GRÁFICA DE RAMA 2-3")
plt.show()
print(f"PUNTO ALEATORIO GENERADO EN ({i}, {j})")

###################################################

M = np.full((nc,nf,3), [255,255,255])

#RAMA 3-4

for x,y in zip(range(i,nf,1),range(j,nc,1)):
   M[x,y] = color
for x,y in zip(range(i,nf,1),range(j,-1,-1)):
   M[x,y] = color

plt.imshow(M)
plt.title("GRÁFICA DE RAMA 3-4")
plt.show()
print(f"PUNTO ALEATORIO GENERADO EN ({i}, {j})")

###################################################

M = np.full((nc,nf,3), [255,255,255])

#RAMA 1-4
for x,y in zip(range(i,-1,-1),range(j,nc,1)):
   M[x,y] = color
for x,y in zip(range(i,nf,1),range(j,nc,1)):
   M[x,y] = color

plt.imshow(M)
plt.title("GRÁFICA DE RAMA 1-4")
plt.show()
print(f"PUNTO ALEATORIO GENERADO EN ({i}, {j})")
```
